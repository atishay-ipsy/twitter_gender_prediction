Create dictionary of 10 commmon color names with key as the hex value of the colors and value as the color name

```python
common_colors_hex_to_names = {'#000000': 'black',
 '#000080': 'navy',
 '#0000ff': 'blue',
 '#008000': 'green',
 '#008080': 'teal',
 '#00ff00': 'lime',
 '#00ffff': 'aqua',
 '#800000': 'maroon',
 '#800080': 'purple',
 '#808000': 'olive',
 '#808080': 'grey',
 '#c0c0c0': 'silver',
 '#ff0000': 'red',
 '#ff00ff': 'fuchsia',
 '#ffff00': 'yellow',
 '#ffffff': 'white'}
```

```python
def hex_to_rgb(hex_color):
    '''
    Takes hex color as input and return RGB values of the color
    '''
    hexC = hex_color.strip('#')
    z = []
    for j in (0, 2 ,4):
        c = int(hexC[j:j+2], 16)
        z.append(c)
    return z
```

```python
def get_colour_name(rgb_triplet):
    '''
    Take rgb color as input and returns the name of the color which is closest to it from the 10 most common color.
    Closest color is defined as one with the least euclidiean distance between the rgb values
    '''
    min_colours = {}
    for key, name in common_colors_hex_to_names.items():
        r_c, g_c, b_c = hex_to_rgb(key)
        rd = (r_c - rgb_triplet[0]) ** 2
        gd = (g_c - rgb_triplet[1]) ** 2
        bd = (b_c - rgb_triplet[2]) ** 2
        min_colours[(rd + gd + bd)] = name
    return min_colours[min(min_colours.keys())]
```
